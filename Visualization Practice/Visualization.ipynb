```python
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
#Readding Shapefile
world=gpd.read_file('world.shp')
world= world[['NAME','geometry']]
```

```python
#Calculating Area
world['AREA']= world.area
```

```python
#Removing some areas from Geopandas
world= world[world['NAME'] != 'Antarctica']
world.plot()
```

```python
#changing the projection system 
world.to_crs(epsg= 3857, inplace=True)
```

```python
#Visualizing maps
world.plot(cmap='hsv', column='AREA' , legend=True, legend_kwds= {'label': 'Area of the country (Sq. Km)'}, figsize=(20,20))
```

```python
#Resizing the legend
fig, ax= plt.subplots(figsize=(20,20))
divider= make_axes_locatable(ax)
cax= divider.append_axes('right', size= '5%', pad=0.1)
world.plot(ax=ax, cmap='hsv', column="AREA", legend= True, legend_kwds={'label':'Area of Country (Sq. Km)'}, cax=cax)
```
